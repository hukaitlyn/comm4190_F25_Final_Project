### Usage Scenario 1: First-Time Runner

### Short Description
A complete beginner with no running experience uses RunMate to assess their fitness level, set realistic goals, and receive a personalized 8-week couch-to-5K training plan.

### Narrative of Scenario
The user has never run consistently but wants to complete their first 5K race in two months. They open RunMate and begin a conversational onboarding. The LLM asks about her current activity level, any past running attempts, time availability, and goals. The user tells the LLM their experience. The LLM responds, validates her goal as achievable, and creates a progressive plan starting with walk/jog intervals. The app generates a structured 8-week plan with gradually increasing run intervals, rest days, and weekly check-ins.

### Step by Step Breakdown
- User opens app → LLM greets user warmly and asks permission to learn about their running background
- LLM asks: "Have you run before?" → User responds: "I tried a few times but couldn't keep going"
- LLM asks: "How long can you comfortably jog right now?" → User: "Maybe 5 minutes before I'm exhausted"
- LLM asks: "What's your goal?" → User: "I want to run a 5K in 2 months"
- LLM asks: "How many days per week can you commit?" → User: "3 days, maybe 30 minutes each"
- LLM asks: "Any injuries or health concerns?" → User: "No, just out of shape"
- LLM validates goal → "That's a great goal and totally achievable! Let's build a plan that starts where you are."
- LLM generates plan → Creates Week 1: 3x walk/jog intervals (5 min walk, 1 min jog, repeat)
- LLM explains structure → "We'll gradually increase your running time each week. Rest days are built in."
- User confirms → Plan is saved to user profile

Data Needed:
- User responses to onboarding questions (text)
- Current fitness level (self-assessed: "can jog 5 minutes")
- Time availability (number format: 3 days/week, 30 min/session)
- Goal description (text: "complete a 5K")
- Timeline (date/number: "2 months")
- Health constraints (text or null)
- Provided to LLM as: Structured conversation history + user profile JSON

### Evaluation
Success criteria:
- LLM identifies fitness level correctly ("maybe 5 minutes" → beginner)
- Plan matches constraints (3 days/week, 30 min, 8 weeks)
- Progressive structure 
- Encouraging tone without dismissiveness
  
Complications:
- Unrealistic goals → LLM redirects to safe alternatives
- Vague responses → Ask simpler questions
- Contradictory info → Clarify discrepancies

### Usage Scenario 2: Pre-Race Week Motivation and Taper Guidance

### Short Description
A runner one week before their first 10K race feels anxious and tempted to cram extra training. RunMate provides race-week taper guidance, manages anxiety, and builds confidence through conversational coaching.

### Narrative of Scenario
The user has been training for their 10K race is in 6 days. They messages RunMate: "My race is Saturday and I'm freaking out. Should I do an extra run tomorrow to feel more prepared?" The LLM detects pre-race anxiety and the common mistake of overtraining during taper week. It responds empathetically, validates their nerves as normal, and explains why rest is more important than extra training right now. The LLM walks them through a proper taper week plan: a short easy 2-mile run on Monday, complete rest Tuesday-Thursday, a 15-minute shakeout jog Friday, then race day Saturday. It also provides pre-race mental preparation, discussing race-day logistics, pacing strategy, and what to eat. Throughout the week, RunMate checks in daily with encouragement and reminders, helping the user trust their training and arrive at the start line rested and confident.

### Step by Step Breakdown
- User messages with anxiety → "My race is Saturday and I'm freaking out"
- LLM detects pre-race stress → Acknowledges feelings: "Race week nerves are totally normal!"
- User asks → "Should I do an extra run tomorrow to feel more prepared?"
- LLM recognizes taper mistake → "I understand wanting to do more, but here's why rest helps..."
- LLM explains taper science → "Your fitness is already built. Now your body needs to be fresh for race day"
- LLM proposes taper plan → "Let's do: Mon 2mi easy, Tue-Thu rest, Fri 15min shakeout"
- User worries → "Won't I lose fitness if I rest that much?"
- LLM reassures → "No—you build fitness in training, you reveal it with rest"
- LLM shifts to race prep → "Let's talk race day strategy. What's your goal pace?"
- User responds → "I want to finish under 60 minutes"
- LLM calculates → "That's 9:40/mile. Start at 10:00/mile pace for first 2 miles..."
- LLM covers logistics → Discusses warmup, what to eat night before, race morning routine
- User feels calmer → "Okay, I trust the plan. Thanks"
- LLM schedules check-ins → "I'll check in daily this week. You've got this!"
- Daily check-ins throughout week → Encouragement, logistics reminders, confidence building
  
Data Needed:
- User's current emotional state (text: "freaking out")
- Race date and details (date: Nov 12, distance: 10K, goal: sub-60min)
- Complete training history (12 weeks of runs, progression, recent performance)
- User's typical paces from training logs
- Current date (6 days before race)
- Provided to LLM as: Conversation history + race details + training summary + calendar

### Evaluation
Success Criteria:
- Detects pre-race anxiety and validates feelings
- Prevents common taper mistakes (overtraining)
- Provides science-based taper plan (progressive rest)
- Generates realistic pacing strategy based on training data
- Reduces user anxiety (sentiment shifts from anxious to confident)

Complications:
- User insists on extra training → Explain risks of showing up tired vs fresh
- Unrealistic race goal → Gently adjust expectations based on training paces
- Race day morning panic → Provide calming checklist and routine
- User reports illness/injury during taper → Adjust expectations, prioritize health over race
