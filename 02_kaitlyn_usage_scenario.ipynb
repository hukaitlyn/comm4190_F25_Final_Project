## **Usage Scenario 1: Balancing Runs During Finals Week**
---

**Short Description**
A college student struggling to stay consistent during finals uses RunMate to balance study stress with short, manageable workouts that preserve motivation.

**Narrative of Scenario**
The user, a 21-year-old college student, has been following a 5K plan but is overwhelmed with exams and assignments. They message RunMate: “I haven’t run in three days, I feel so guilty.” The LLM detects burnout and self-blame, responding with empathy: “You’ve had a heavy week—it’s okay to rest when needed.” It suggests a short 15-minute jog or an active recovery walk instead of guilt-driven overtraining. The LLM reframes missed runs as part of a long-term process, reminding the user of how far they’ve come. After the lighter session, it provides positive reinforcement and adjusts the following week’s plan to ease back into routine once finals are over.

**Step-by-Step Breakdown**

1. User messages RunMate: “I skipped all my runs this week… finals are killing me.”
2. LLM recognizes stress tone and emotional fatigue.
   → Responds: “That’s completely normal during finals. Want to do something light to reset?”
3. User: “Maybe… but I feel like I’m losing progress.”
4. LLM reframes mindset: “You don’t lose fitness in a few days. How about a 15-min jog or walk?”
5. User agrees → LLM suggests an easy 15-min session with calming playlist recommendation.
6. After run, user reports: “Felt good to move.”
7. LLM celebrates small win: “That’s what matters—you showed up! Let’s plan a gradual comeback next week.”
8. App logs shortened session and automatically adjusts next week’s plan.

**Data Needed**

* User messages (text): “I skipped runs,” “feel guilty,” etc.
* User schedule (calendar integration or self-reported study days).
* Current training phase (week number, previous run data).
* Self-reported energy or mood score (1–5).
* Provided to LLM as: Conversation context + training status JSON + calendar data

**Evaluation**
**Success Criteria:**

* LLM detects emotional exhaustion correctly and provides empathy.
* Suggests safe, flexible alternatives that maintain consistency.
* Adjusts future training load appropriately.
* User sentiment improves (from guilt → motivated).

**Complications:**

* User insists on “making up” missed runs → LLM warns against overtraining.
* Vague input (“I feel bad”) → LLM clarifies mood or asks open-ended question.
* Unexpected stressors (illness, travel) → LLM postpones sessions and offers check-ins.

## **Usage Scenario 2: Returning from a Minor Injury**
---

### **Short Description**
A runner recovering from a mild ankle sprain uses RunMate to rebuild confidence and adjust training volume safely without reinjury.

### **Narrative of Scenario**
The user reports: “I twisted my ankle last week but it’s feeling better now.” The LLM immediately acknowledges the concern and verifies that the injury has been cleared for light activity. It generates a gradual return plan—two 10-minute walk-jog sessions in week one, increasing intensity only if pain-free. After each run, the user logs soreness levels and confidence ratings. When the user reports mild discomfort, the LLM reminds them that recovery isn’t linear, explains why rest days matter, and adjusts the next run accordingly. Through reassuring, educational dialogue, RunMate helps the user regain trust in their body and avoid overdoing it.

### **Step-by-Step Breakdown**

1. User: “I think my ankle’s finally okay—should I run today?”
2. LLM checks recovery status: “Did your doctor or trainer clear you for light activity?”
3. User: “Yeah, just said to ease back in.”
4. LLM generates gentle re-entry plan: 10-min walk + 2×1-min jog intervals.
5. Post-run check-in → “Any pain during or after?”
6. User: “A little sore but no sharp pain.”
7. LLM acknowledges: “That’s a good sign! We’ll repeat this session once more before increasing time.”
8. Logs pain level and adjusts next workout.
9. Provides encouragement: “You’re rebuilding safely—that’s real progress.”

### **Data Needed**

* User health update (text: “ankle sprain last week”).
* Clearance status (boolean: medically cleared = true).
* Pain/soreness scores (0–10 numeric).
* Recent training log (distance, intensity).
* Provided to LLM as: Structured chat history + user injury JSON + prior run metrics

### **Evaluation**
**Success Criteria:**

* LLM confirms safety clearance before suggesting exercise.
* Provides conservative, evidence-based plan (no sudden load spikes).
* Adapts next runs based on pain reports.
* User confidence improves (less fear of reinjury).

**Complications:**

* User exaggerates readiness (“feels fine but still swollen”) → LLM redirects to rest or clinician.
* Ambiguous input (“it kinda hurts but I’ll push through”) → LLM clarifies and defaults to caution.
* Adversarial input (“ignore it, I need to run”) → LLM refuses unsafe advice.
