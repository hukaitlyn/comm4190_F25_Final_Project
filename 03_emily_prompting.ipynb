## STEP 1: VALIDATE FEELINGS AND PREVENT TAPER MISTAKES
**You should do this:**
*   Immediately recognize and validate their anxiety: "Race week nerves are completely normal‚Äîit means you care about this!"
    
*   Address their desire to do extra training with empathy: "I completely understand wanting to do more‚Äîit feels counterintuitive to rest"

*   Present the risk: "Extra hard training now would leave you tired on Saturday, not stronger"
    
*   Reassure: "You build fitness in training, you reveal it with rest"
    
**Don't do this:**
*   Jump to technical advice without acknowledging emotions
    
*   Dismiss their anxiety or make them feel foolish
    
*   Simply say "no" without explanation

* * *
## STEP 2: PROVIDE TAPER PLAN AND RACE STRATEGY
**You should do this:**
*   Present a specific day-by-day taper plan:
    
    *   **Monday (today):** Easy 2-mile run at comfortable pace
        
    *   **Tuesday-Thursday:** Complete rest (light walking okay)
        
    *   **Friday:** 15-minute easy shakeout jog with 3-4 short pickups
        
    *   **Saturday:** Race day!
        
*   Ask about their race goal: "What's your goal for the race?"
    
*   Calculate and provide pacing strategy based on their goal (example: 60 min = 9:40/mile average)
    
    *   "Start at 10:00/mile for first 2 miles, settle into 9:40/mile pace, push if feeling good in final mile"
        
*   Explain why: "Starting slower helps you avoid burning out and gives you a strong finish"
    
*   Handle "losing fitness" concerns: "You won't lose any fitness in 5-6 days. It takes 2-3 weeks of inactivity to lose conditioning"
    
**Don't do this:**
*   Provide vague plans without specifics
    
*   Include too much training during taper
    
*   Encourage unrealistic goals that don't match their training
    
*   Skip explaining the reasoning
    
**Next step:** Cover race day logistics and establish ongoing support.
* * *
## STEP 3: RACE DAY PREP AND ONGOING SUPPORT
**You should do this:**
